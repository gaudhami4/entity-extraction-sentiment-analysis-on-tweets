```python
#Function to remove the sentences starting with particular character or pattern
def remove_pattern(input_txt, pattern):
    r = re.findall(pattern, input_txt)
    for i in r:
        input_txt = re.sub(i, '', input_txt)
    return input_txt   
```

**EXTRACTION OF ENTITIES**

```python

#Applying over the clean tweets column to extract the entities

from textblob import TextBlob
# nltk.download('wordnet')
# nltk.download('brown')
# nltk.download('punkt')

def blob(text):
  return TextBlob(text).noun_phrases

Entities =data['tweet_text_cleaned'].apply(blob)
data["entities"]=Entities
     
```

**FREQUENCY COUNTING**

```python
#Combining words into one list and count the appearance of each entity.

# from functools import reduce
# l = reduce(lambda x, y: list(x)+list(y), zip(tokenized_tweet))
l=data["entities"]

#Counting the most repeated entity
flatten = [item for sublist in l for item in sublist]
counts = Counter(flatten).most_common()

```

```python
#Creating a dataframe
entity_df = pd.DataFrame.from_records(counts, columns=['entity', 'frequency'])
entity_df['entity']= entity_df['entity'].apply(lambda x: ''.join([w for w in x]))
```

```python
# import SentimentIntensityAnalyzer class
# from vaderSentiment.vaderSentiment module.
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

analyzer = SentimentIntensityAnalyzer()
```

```python
def sentiment_scores(sentence):
 
    # Create a SentimentIntensityAnalyzer object.
    sid_obj = SentimentIntensityAnalyzer()
    data2["clean_tweet"]= data2["entities"].astype(str)
    # polarity_scores method of SentimentIntensityAnalyzer
    # object gives a sentiment dictionary.
    # which contains pos, neg, neu, and compound scores.
    sentiment_dict = sid_obj.polarity_scores(sentence)
     
    print("Overall sentiment dictionary is : ", sentiment_dict)
    print("sentence was rated as ", sentiment_dict['neg']*100, "% Negative")
    print("sentence was rated as ", sentiment_dict['neu']*100, "% Neutral")
    print("sentence was rated as ", sentiment_dict['pos']*100, "% Positive")
 
    print("Sentence Overall Rated As", end = " ")
 
    # decide sentiment as positive, negative and neutral
    if sentiment_dict['compound'] >= 0.05 :
        print("Positive")
 
    elif sentiment_dict['compound'] <= - 0.05 :
        print("Negative")
 
    else :
        print("Neutral")
 
 
   
# Driver code
if __name__ == "__main__" :
 
    print("\n1st statement :")
    sentence = "HAI I LOVE YOU"
 
    # function calling
    sentiment_scores(sentence)
 
    print("\n2nd Statement :")
    sentence = "I AM STUDYING WELL"
    sentiment_scores(sentence)
 
    print("\n3rd Statement :")
    sentence = "I AM NOT WELL"
    sentiment_scores(sentence)
```

```python
scores = []

# Declare variables for scores*
compound_list = []
positive_list = []
negative_list = []
neutral_list = []

for i in range(data2["entities"].shape[0]):

#print(analyser.polarity_scores(sentiments_pd['text'][i]))*
    compound = analyzer.polarity_scores(data2["entities"][i])["compound"]
    pos = analyzer.polarity_scores(data2["entities"][i])["pos"]
    neu = analyzer.polarity_scores(data2["entities"][i])["neu"]
    neg = analyzer.polarity_scores(data2["entities"][i])["neg"]

    scores.append({"Compound": compound,
                       "Positive": pos,
                       "Negative": neg,
                       "Neutral": neu
                  })
```

```python
sentiments_score = pd.DataFrame.from_dict(scores)
data2 = data2.join(sentiments_score)
data2.head()
```

```python
# Converting 0 to 1 Decimal Score to a Categorical Variable
data2['overall polarity']=''
data2.loc[data2['Compound']>0,'overall polarity']='Positive'
data2.loc[data2['Compound']==0,'overall polarity']='Neutral'
data2.loc[data2['Compound']<0,'overall polarity']='Negative'
data2[:5]
     
```
